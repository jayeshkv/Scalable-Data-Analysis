```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import palettable.colorbrewer as bw
import seaborn 
seaborn.set(color_codes=True) 

fileNames = [];

stateCount = pd.DataFrame();
UgStateWise = pd.DataFrame();
avgFacSalary_State = pd.DataFrame();
adm_Rate = pd.DataFrame();
UgLowInc = pd.DataFrame()
UgMidClass = pd.DataFrame();
UgHiClass = pd.DataFrame();
csPassouts = pd.DataFrame();
univFee = pd.DataFrame();

for x in range(1997,2014):
    x = str(x)
    fileNames.append("MERGED"+x+"_PP");

for files in fileNames:
    csv = pd.read_csv(files+".csv")
    csv = csv.convert_objects(convert_numeric=True)
    
    csv['HiClass'] = csv[['INC_PCT_H1','INC_PCT_H2']].sum(axis=1,numeric_only=True) 
    csv['MidClass'] = csv['INC_PCT_M1'] + csv['INC_PCT_M2']
  
    stateCount[files] = csv["STABBR"].value_counts(); #statewise school count

    UgStateWise[files] = csv.groupby(by=['STABBR'])['UGDS'].sum() #statewise UG count
    avgFacSalary_State[files] = np.round(csv.groupby(by=['STABBR'])['AVGFACSAL'].mean()) # statewise salary mean of faculty
    
    adm_Rate[files] = csv.groupby(by=['STABBR'])['ADM_RATE_ALL'].mean()#admission rate no:of applied / no:of admitted  | from 2001
    UgLowInc[files] = csv.groupby(by=["STABBR"])['INC_PCT_LO'].mean().fillna(0) #low income candidates - $0-$30,000 works for all years
    UgMidClass[files] = csv.groupby(by=["STABBR"])['MidClass'].mean().fillna(0) #middle class candidates - $30,000 - 75k works for all years
    UgHiClass[files] = csv.groupby(by=['STABBR'])['HiClass'].mean().fillna(0) #high class candidates 75k+ | works for all years 
    csPassouts[files] = csv.groupby(['STABBR'])['PCIP15'].mean()#https://nces.ed.gov/ipeds/cipcode/browse.aspx?y=55 15 points to Tech Enginnering
    univFee[files] = csv.groupby(['STABBR'])['COSTT4_P'].mean()

    #UgLowInc.dropna()
#UgLowInc
#csv = pd.read_csv('MERGED1996_PP.csv')
#stateCount = csv["STABBR"].value_counts()#statewise univ counts
#UgStateWise = csv.groupby(by=['STABBR'])['UGDS'].sum() #statewise UG count
#avgFacSalary_State = csv.dropna(subset=['AVGFACSAL']) #removing all the null values which is already removed by groupby
#avgFacSalary_State = np.round(avgFacSalary_State.groupby(by=['STABBR'])['AVGFACSAL'].mean()) # statewise salary mean of faculty

#adm_Rate #works only for 2001 & 2014 | admission rate 1 being easy 
#Pct_Pell_Grant # scholarship data available after 2008
#INC_PCT_H1 and others between 2010 & 2014
```
