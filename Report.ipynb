## Visualizing various parameters using the "College Scorecard Dataset" from US Dept of Education

<h3>Overview</h3>

We are trying to examine the dataset that is available from the US department of Education website, the dataset from the website contains data from 1996 - 2013 in the form of csv split according to the year. The excel files that we are trying to examine contains more than 1200 columns and some values are availabe for certain years whereas some are available for all years. Some of the columns are "INSTNM" which represents the institution name and "STABBR" & "CITY" which represents the state and city at which the institution is located in and so on...

<h3>Discussion</h3>

Since the data that we got is in csv format, the data needs to be cleaned and typecasted. For this process we use python since it has a lot of pre built functions that could be used for data analysis and data visualization. It has the library called pandas that can be used for data cleaning, data retrieval and matplot lib is used for plotting the graphs. Seaborn is used to provide color palettes for the graphs to make it look good.
